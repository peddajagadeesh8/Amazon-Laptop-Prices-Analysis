```python
import matplotlib.pyplot as plt
#bar
plt.figure(figsize=(10, 6))
plt.bar(data['Location'], data['Salaries Reported'])
plt.xlabel('Location')
plt.ylabel('Salaries Reported')
plt.title('Bar Plot of Salaries Reported by Location')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
data['Job Title'].value_counts().sort_index().plot.bar()
plt.xlabel('Job Title', fontsize = 25)
plt.ylabel('Count', fontsize = 25)
plt.title('Bar Plot of Job Titles', fontsize = 25)
plt.xticks(rotation=80, fontsize=16)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Assigning data to df variable
df = pd.DataFrame(data)
# Creating a pie chart for car owners
df['Job Title'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Job Titles',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating bar chart for top 10 cars sold
plt.figure(figsize=(6,4))
df['Company Name'].value_counts().head(10).plot(kind='bar',color='#FF0000',fontsize=10,edgecolor='black',width=0.8)
plt.xlabel('Cars',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.title('Top 10 Cars sold',weight='bold',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
data.isnull().sum()
```

```python
# Creating bar chart of cars sold in a particular year
plt.figure(figsize=(8,4))
df['Job Title'].value_counts().sort_index().plot(kind='bar',color='#2ca02c',edgecolor='black',width=0.8)
plt.title('Count of Cars by Year',weight='bold',fontsize=14)
plt.xlabel('Year',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
```

```python
# Creating pie chart for dealers
colors = ['gold', 'yellowgreen', 'lightcoral']
df['Location'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=colors,startangle=90)
plt.title('Dealers',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))
plt.xticks(rotation=140)
plt.show()
```

```python
# Group by 'Location' and count the unique 'Company' values in each 'Location'
company_counts = data.groupby('Location')['Company Name'].nunique()

plt.figure(figsize=(15, 8))
company_counts.plot(kind='bar')
plt.xlabel('Location')
plt.ylabel('Number of Companies')
plt.title('Number of Companies in Each Location')
plt.show()
```
